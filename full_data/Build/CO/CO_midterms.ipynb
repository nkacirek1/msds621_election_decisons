```python
import pandas as pd
```

```python
mid = pd.read_excel('/Users/nicolekacirek/Downloads/2014GeneralPrecinctResults.xlsx')
```

```python
districts = ['United States Representative - District 1', 'United States Representative - District 2',
             'United States Representative - District 3', 'United States Representative - District 4',
             'United States Representative - District 5', 'United States Representative - District 6',
             'United States Representative - District 7', 'United States Senator']
```

```python
d = mid[mid['Office/Issue/Judgeship'].isin(districts)]
```

```python
party = ['Republican Party', 'Democratic Party']
d = d[d.Party.isin(party)]
```

```python
d = d.drop(columns=['Year', 'Election Type', 'Yes Votes', 'No Votes'])
```

```python
grouped_d = d.groupby(['State', 'County', 'Precinct', 'Office/Issue/Judgeship', 'Party']).sum()
```

```python
with_sums = d.copy()
```

```python
use = with_sums.join(grouped_d, on=['State', 'County', 'Precinct', 'Office/Issue/Judgeship', 'Party'], rsuffix='_r')
```

```python
dmerged = pd.merge(use, use, on=['State', 'County', 'Precinct', 'Office/Issue/Judgeship'])
```

```python
dmerged
```

```python
dmerged = dmerged.drop(columns=['Candidate_x', 'Candidate_y', 'Candidate Votes_x', 'Candidate Votes_y'])
```

```python
dmerged = dmerged[dmerged['Party_x'] != dmerged['Party_y']] # removed rows with matching partys
```

```python
dmerged = dmerged.drop_duplicates()
```

```python
dmerged = dmerged[dmerged['Party_x'] == 'Democratic Party']# keep only row with dem first
```

```python
dmerged
```

```python
dmerged.columns = ['STATE', 'COUNTY', 'PRECINCT', 'DISTRICT','Party_x', 'DEM', 'Party_y', 'REP']
```

```python
dmerged = dmerged.drop(columns=['Party_x', 'Party_y'])
dmerged.STATE = 'CO'
```

```python
dmerged.COUNTY = dmerged.COUNTY.str.upper()
```

```python
dmerged.PRECINCT = dmerged.PRECINCT.astype(str).str.upper()
```

```python
dmerged = dmerged.drop_duplicates()
```

```python
dmerged
```

```python
d = mid[mid['Office/Issue/Judgeship'].isin(districts)]
```

# Debug

```python
dist = d[d['Office/Issue/Judgeship'] == 'United States Representative - District 1'][['County', 'Precinct', 'Office/Issue/Judgeship']]
```

```python
dist = dist.drop_duplicates()
```

```python
dist.County = dist.County.str.upper()
dist.Precinct = dist.Precinct.astype(str).str.upper()
```

```python
d_final = pd.merge(dist, dmerged, how='left', left_on=['County','Precinct', 'Office/Issue/Judgeship'], right_on = ['COUNTY','PRECINCT','DISTRICT'])[['STATE', 'COUNTY', 'PRECINCT','DISTRICT','DEM','REP']]
```

```python
d_final = d_final[d_final.DISTRICT == 'United States Representative - District 1']
```

```python
d_final = d_final.drop(columns=['DISTRICT'])
```

# End Debug

```python
for i in range(len(districts)):
    dist = d[d['Office/Issue/Judgeship'] == districts[i]][['County', 'Precinct', 'Office/Issue/Judgeship']]
    dist = dist.drop_duplicates()
    dist.County = dist.County.str.upper()
    dist.Precinct = dist.Precinct.astype(str).str.upper()
    d_final = pd.merge(dist, dmerged, how='left', left_on=['County','Precinct', 'Office/Issue/Judgeship'], right_on = ['COUNTY','PRECINCT','DISTRICT'])[['STATE', 'COUNTY', 'PRECINCT','DISTRICT','DEM','REP']]
    d_final = d_final[d_final.DISTRICT == districts[i]]
    d_final = d_final.drop(columns=['DISTRICT'])
    filename = 'D%d_2014.csv' % (i+1)
    d_final.to_csv(filename, index=False)
```
