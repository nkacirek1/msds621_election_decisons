```python
import pandas as pd
```

```python
prez = pd.read_excel('/Users/nicolekacirek/Downloads/2012GeneralPrecinctLevelResults.xlsx')
mid = pd.read_excel('/Users/nicolekacirek/Downloads/2014GeneralPrecinctResults.xlsx')
```

```python
candidates = ['Barack Obama / Joe Biden', 'Mitt Romney / Paul Ryan']
```

```python
prez = prez[prez.Candidate.isin(candidates)]
```

```python
prez = prez.drop(columns=['Year', 'Election Type', 'Office/Issue/Judgeship', 'Candidate', 'Yes Votes', 'No Votes'])
```

```python
merged = pd.merge(prez, prez, on=['State', 'County', 'Precinct'])
```

```python
merged = merged[merged['Party_x'] != merged['Party_y']] # removed rows with matching partys
merged = merged[merged['Party_x'] == 'Democratic Party']# keep only row with dem first
```

```python
merged.columns = ['STATE', 'COUNTY', 'PRECINCT', 'Party_x', 'PAST_DEM', 'Party_y', 'PAST_REP']
```

```python
merged = merged.drop(columns=['Party_x', 'Party_y'])
```

```python
merged.STATE = 'CO'
```

```python
merged.COUNTY = merged.COUNTY.str.upper()
```

```python
merged
```

```python
merged.PRECINCT = merged.PRECINCT.astype(str).str.upper()
```

Need to get the matching from county to districts

```python
mid
```

```python
districts = ['United States Representative - District 1', 'United States Representative - District 2',
             'United States Representative - District 3', 'United States Representative - District 4',
             'United States Representative - District 5', 'United States Representative - District 6',
             'United States Representative - District 7', 'United States Senator']
```

```python
d = mid[mid['Office/Issue/Judgeship'].isin(districts)]
```

```python
d
```

```python
for i in range(len(districts)):
    dist = d[d['Office/Issue/Judgeship'] == districts[i]][['County', 'Precinct']]
    dist = dist.drop_duplicates()
    dist.County = dist.County.str.upper()
    dist.Precinct = dist.Precinct.astype(str).str.upper()
    d_final = pd.merge(dist, merged, how='left', left_on=['County','Precinct'], right_on = ['COUNTY','PRECINCT'])[['STATE', 'COUNTY', 'PRECINCT','PAST_DEM','PAST_REP']]
    filename = 'D%d_2012.csv' % (i+1)
    d_final.to_csv(filename, index=False)
```
